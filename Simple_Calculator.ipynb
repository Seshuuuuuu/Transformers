```python
from transformers import pipeline, T5Tokenizer, T5ForConditionalGeneration
from langchain.agents import initialize_agent, Tool
from langchain_huggingface import HuggingFacePipeline
from langchain.agents.agent_types import AgentType
import torch
from langchain_community.llms import HuggingFaceHub
```

```python
# Load the instruction-tuned Flan-T5 model (small)
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-base")
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-base")
```

```python
# Set up the generation pipeline
pipe = pipeline("text2text-generation", model=model, tokenizer=tokenizer)
```

```python
# Custom calculator tool (simple eval, use safely)
def calculator(expression: str) -> str:
    try:
        return str(eval(expression))
    except:
        return "Invalid math expression."
```

```python
# Ask a question
user_question = input("Enter the Math Expression..")

# Step 1: Ask the model to analyze the math expression
prompt = f"""
You are a smart assistant. You can break down math problems and solve them step by step.

Question: What is {user_question}?
Explain your reasoning step-by-step first.
Then write the final answer.
"""

# Step 2: Run the model to generate reasoning
response = pipe(prompt, max_new_tokens=150)[0]['generated_text']
print("🧠 AI Reasoning:\n", response)
```

```python
# Step 3: Let the calculator solve it (simulate tool use)
calc_result = calculator(user_question)
print("\n🧮 Tool Output (Calculator):", calc_result)
```
